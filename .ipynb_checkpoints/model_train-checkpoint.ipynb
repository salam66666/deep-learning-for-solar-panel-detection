## **Part 1: Importing libraries and setting parameters**

```python
import os
import json
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import Sequence
from matplotlib import pyplot as plt

# route
image_dir = r'E:\CDUT\English\7 Term\Project\model\archive\dataset\images'
annotation_dir = r'E:\CDUT\English\7 Term\Project\model\archive\dataset\annotations'

# Image size and batch size
IMG_HEIGHT = 512
IMG_WIDTH = 640
BATCH_SIZE = 16
```

## **Part 2: Data load function: load images and all tags**

```python

def load_dataset(image_dir, annotation_dir):
    images = []  
    all_bboxes = [] 
    all_labels = []
    
    # 遍历图片和JSON文件
    for filename in os.listdir(image_dir):
        if filename.endswith(".jpg") or filename.endswith(".png"):
            image_path = os.path.join(image_dir, filename)
            annotation_path = os.path.join(annotation_dir, filename.replace('.jpg', '.json').replace('.png', '.json'))
            
            # 加载图片并缩放到指定大小
            img = load_img(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH))
            img_array = img_to_array(img) / 255.0  # 归一化
            images.append(img_array)
            
            # 加载JSON标注
            with open(annotation_path, 'r') as f:
                annotation = json.load(f)
                instances = annotation.get("instances", [])
                
                bboxes = []
                labels = []
                
                for instance in instances:
                    # 获取四个角点
                    corners = instance.get("corners", [])
                    if len(corners) == 4:  # 确保有4个角点
                        # 将角点转换为numpy数组
                        points = np.array([[corner["x"], corner["y"]] for corner in corners], dtype=np.float32)
                        
                        # 计算四个角点的凸包
                        hull = cv2.convexHull(points)
                        
                        # 计算凸包的最小外接矩形
                        rect = cv2.minAreaRect(hull)  # 返回的是(中心, (宽, 高), 角度)
                        
                        # 获取外接矩形的四个角点
                        box = cv2.boxPoints(rect)  # 直接返回浮点型坐标
                        
                        # 将四个角点的坐标作为边界框
                        bboxes.append(box)
                    
                    # 获取标签
                    labels.append(1 if instance.get("defected_modules", False) else 0)
                
                all_bboxes.append(bboxes)
                all_labels.append(labels)
    
    return np.array(images), all_bboxes, all_labels

# 调用数据加载函数
images, bboxes, labels = load_dataset(image_dir, annotation_dir)

# 检查数据形状
print(f"图片数据形状: {images.shape}")
print(f"边界框数量示例（前1张图像）: {len(bboxes[0])}")
print(f"标签示例（前1张图像的所有标签）: {labels[0]}")
```

```python
def visualize_annotations(images, corners_list, img_idx=0):
    img = images[img_idx]
    corners = corners_list[img_idx]
    
    img_height, img_width, _ = img.shape
    
    # 创建一个图像
    plt.imshow(img)
    
    for corner in corners:
        if isinstance(corner, list) and len(corner) == 4:
            x = [point['x'] for point in corner]
            y = [point['y'] for point in corner]
            
            # 裁剪坐标，使其不超出图像范围
            x = np.clip(x, 0, img_width - 1)
            y = np.clip(y, 0, img_height - 1)
            
            # 连接四个角点，绘制矩形框
            plt.plot(x + [x[0]], y + [y[0]], color='red', lw=2)  # 连接四个角点
            
            # 标注中心点
            center = np.mean(corner, axis=0)
            center[0] = np.clip(center[0], 0, img_width - 1)  # 确保中心点在图像范围内
            center[1] = np.clip(center[1], 0, img_height - 1)  # 确保中心点在图像范围内
            plt.scatter(center[0], center[1], color='blue', marker='x', s=100, label="Center")
    
    # 设置坐标轴范围
    plt.xlim(0, img_width)
    plt.ylim(img_height, 0)  # Y轴反向
    plt.show()

# 加载数据集
train_images, train_labels, train_corners = load_dataset(image_dir, annotation_dir)

# 可视化第一个训练图片的全部标注
visualize_annotations(train_images, train_corners, img_idx=0)
```

## **Part 3: Divide the data set into a training set and a validation set**

```python

# 划分数据集为训练集和验证集
x_train, x_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)

# 打印划分后的数据形状
print(f"训练集图片数量: {x_train.shape[0]}, 验证集图片数量: {x_val.shape[0]}")
```

## **Part 4: Define data generator**

```python
class DataGenerator(Sequence):
    def __init__(self, images, labels, batch_size):
        """
        初始化数据生成器
        :param images: 图像数据数组 (num_samples, height, width, channels)
        :param labels: 标签数组（嵌套列表，每张图像的多个标签）
        :param batch_size: 批次大小
        """
        self.images = images
        self.labels = labels
        self.batch_size = batch_size
        self.indexes = np.arange(len(images))
    
    def __len__(self):
        """
        返回每个 epoch 中的批次数量
        """
        return int(np.ceil(len(self.images) / self.batch_size))
    
    def __getitem__(self, index):
        """
        获取一个批次的数据
        """
        batch_indexes = self.indexes[index * self.batch_size:(index + 1) * self.batch_size]
        batch_images = self.images[batch_indexes]
        
        # 处理嵌套标签列表
        batch_labels = [self.labels[i] for i in batch_indexes]
        
        # 如果需要，将标签展平或转换为合适的模型格式
        # 例如：可以用多热编码表示多个缺陷模块的情况
        max_instances = max(len(lbl) for lbl in batch_labels)  # 当前批次中最多的标签数
        batch_labels_padded = np.zeros((len(batch_labels), max_instances), dtype=int)
        for i, lbl in enumerate(batch_labels):
            batch_labels_padded[i, :len(lbl)] = lbl
        
        return batch_images, batch_labels_padded
    
    def on_epoch_end(self):
        """
        每个 epoch 后打乱数据
        """
        np.random.shuffle(self.indexes)


# 创建训练集和验证集的生成器
train_gen = DataGenerator(x_train, y_train, BATCH_SIZE)
val_gen = DataGenerator(x_val, y_val, BATCH_SIZE)
```

## **Part 5: Building and training the model**

```python
# 模型构建函数
def build_model(input_shape):
    model = models.Sequential()
    
    # 卷积层 1
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
    model.add(layers.MaxPooling2D((2, 2)))
    
    # 卷积层 2
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    
    # 卷积层 3
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    
    # 展平层
    model.add(layers.Flatten())
    
    # 全连接层
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(1, activation='sigmoid'))  # 假设是二分类任务
    
    return model

# 调用模型构建函数
model = build_model((IMG_HEIGHT, IMG_WIDTH, 3))

# 打印模型结构
model.summary()

# 编译模型
model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
              loss='binary_crossentropy',  # 二分类任务
              metrics=['accuracy'])

print("模型编译完成")

# 训练模型
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,  # 可根据需求调整
    verbose=1
)

print("模型训练完成")
```

## **Part 6: Drawing training curve and saving model**

```python
# 绘制训练过程曲线
def plot_training_history(history):
    plt.figure(figsize=(12, 4))
    
    # 损失图
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend()
    plt.title('Loss')
    
    # 准确率图
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.legend()
    plt.title('Accuracy')
    
    plt.show()

# 绘制训练曲线
plot_training_history(history)

# 保存模型
model.save("solar_cell_model.h5")
print("model save as solar_cell_model.h5")
```
